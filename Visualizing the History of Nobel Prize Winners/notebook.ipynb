The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
```

```python
# uploading the dataset and transforming it into DataFrame
df = pd.read_csv('data/nobel.csv')
```

```python
# reading first 5 rows of the data
df.head()
```

```python
# What is the most commonly awarded gender and birth country?
top_gender = df['sex'].mode()[0]

top_country = df['birth_country'].mode()[0]

print(top_gender)   # Male
print(top_country)  # United States of America
```

```python
# Creating a decade column used in the next questions
df['decade'] = (df['year'] // 10) * 10
```

```python
# Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
US_born = df[df['birth_country']=='United States of America']

df[['year', 'decade']].head()
total_by_decade = df.groupby('decade').size()
usa_by_decade = US_born.groupby('decade').size()
ratio = usa_by_decade / total_by_decade
max_decade_usa = ratio.idxmax()
print(max_decade_usa)        # 2000 had the highest ratio of US-born Nobel Prize winners 
```

```python
# Which decade and Nobel Prize category combination had the highest proportion of female laureates?
Females = df['sex'] == 'Female'

total_winners = df.groupby(['decade', 'category']).size()

female_winners = df[Females].groupby(['decade', 'category']).size()

female_ratio = female_winners / total_winners

max_index = female_ratio.idxmax()
max_decade, max_category = max_index

max_female_dict = {max_decade: max_category}
print(max_female_dict)  # {2020: 'Literature'}
```

```python
# Who was the first woman to receive a Nobel Prize, and in what category?
Females_df = df[Females]
Females_df.sort_values('year',ascending=True)
first_woman_name = Females_df['full_name'].iloc[0]
first_woman_category = Females_df['category'].iloc[0]
print(first_woman_name)   # Marie Curie, née Sklodowska
print(first_woman_category)   # Physics
```

```python
# Which individuals or organizations have won more than one Nobel Prize throughout the years?
individual_names = df[df['laureate_type'] == 'Individual']['full_name'].dropna()
individual_counts = individual_names.value_counts()
individual_repeats = individual_counts[individual_counts >= 2]

organization_names = df[df['laureate_type'] == 'Organization']['full_name'].dropna()
organization_counts = organization_names.value_counts()
organization_repeats = organization_counts[organization_counts >= 2]

repeat_list = individual_repeats.index.tolist() + organization_repeats.index.tolist()

print(repeat_list)
```
